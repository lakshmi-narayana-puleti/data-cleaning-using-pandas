```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
df
```

```python
#df.drop(columns='Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('[123._/]')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].str.replace('-', '').str.replace('/', '').str.replace('|','')
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '').str.replace('Na--', '')
df
```

```python
df[['Street_Adress', 'State', 'Zip_code']] = df['Address'].str.split(',', expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace({'Y':'Yes', 'N':'No'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y':'Yes', 'N':'No'})
df
```

```python
df['Paying Customer'].fillna(value=df['Paying Customer'].max())
df['Paying Customer'] = df['Paying Customer'].replace('N/a', 'Yes')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: str(x))
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'nan':'No'})
df
```

```python
for i in df.index:
  if df.loc[i, 'Do_Not_Contact'] == 'Yes':
    df.drop(i, inplace=True)
df
```

```python
for i in df.index:
  if df.loc[i, 'Phone_Number'] == '':
    df.drop(i, inplace=True)
df
```
